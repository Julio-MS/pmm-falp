```python
def data(df, categoria):
    selects = []
    size = random.randint(15, 20)
    while len(selects) < size:
        candidate = (random.randint(0, len(df) - 1))
        if candidate not in selects:
            selects.append(candidate)
            id_img = df.loc[candidate, 'image_id']
            #print(id_img)
            utils.copy_dcm(id_img, categoria, len(selects))
    
    print(f'Se guardaron {len(selects)} imágenes de caregoría {categoria}')
```
