## Exploración de imágenes

```python
import os
import pandas as pd
data = '/Users/julio/Documentos-Local/data/DrVinn-Mammo/raw'

findings = pd.read_csv('/Users/julio/Documentos-Local/data/DrVinn-Mammo/metadata/finding_annotations.csv')
df_train = pd.DataFrame(columns=findings.columns)
df_test = pd.DataFrame(columns=findings.columns)

images = os.listdir(data)
for index, image in enumerate(images):

    im_id = image.split('.')[0]
    row = findings.loc[findings['image_id'] == im_id]
    if row['split'].values[0] == 'training':
        df_train = pd.concat([df_train, row], ignore_index=True)
    else:
        df_test = pd.concat([df_test, row], ignore_index=True)
```

## Split de imágenes

```python
import os
import shutil

# Ruta de la carpeta origen
source_dir = '/Users/julio/Documentos-Local/data/DrVinn-Mammo/raw'

# Ruta de la carpeta destino
destination_dir1 = '/Users/julio/Documentos-Local/data/DrVinn-Mammo/split/train'
destination_dir2 = '/Users/julio/Documentos-Local/data/DrVinn-Mammo/split/val'

# Crear la carpeta destino si no existe
os.makedirs(destination_dir1, exist_ok=True)
os.makedirs(destination_dir2, exist_ok=True)

for index, image in df_train.iterrows():
    name = image['image_id']
    file_path = f'{source_dir}/{name}.dicom'
    file_destination = f'{destination_dir1}/{name}.dicom' 
    shutil.copy(file_path, file_destination)

for index, image in df_test.iterrows():
    name = image['image_id']
    file_path = f'{source_dir}/{name}.dicom'
    file_destination = f'{destination_dir2}/{name}.dicom' 
    shutil.copy(file_path, file_destination)
```
